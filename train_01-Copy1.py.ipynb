```python
from sklearn import datasets
from sklearn.utils.validation import check_random_state
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.ensemble import GradientBoostingClassifier  #GBM algorithm
from rgf.sklearn import RGFClassifier
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

name=['Date','Address','Species','Block','Street','Trap','AddressNumberAndStreet','Latitude','AddressAccuracy','NumMosquitos','WnvPresent']

def get_set(filename):
    train_set=pd.read_csv(filename, header=0)


    train = datasets.load_iris()
    train['data'] =np.array([pd.Categorical(train_set['Species']).codes, 
                    [int(train_set['Date'][i].split('-')[1]) * 30 + int(train_set['Date'][i].split('-')[2]) for i in range(len(train_set))],
                    train_set['Latitude'], 
                    train_set['Longitude'], 
                    train_set['AddressAccuracy'], 
                    train_set['NumMosquitos']]).transpose()

    train['DESCR'] = ''
    train['feature_names'] = ['Species', 'Date', 'Latitude', 'Latitude', 'AddressAccuracy', 'NumMosquitos']
    train['target'] = np.array(train_set['WnvPresent'])
    train['target_names'] = train['target']
    return train


train = get_set('train.csv');
#test = get_set('test.csv');


#rng = check_random_state(0)
#perm = rng.permutation(iris.target.size)
#iris.data = iris.data[perm]
#iris.target = iris.target[perm]

rgf = GradientBoostingClassifier(loss='deviance', 
                                 learning_rate=0.0035, 
                                 n_estimators=1000, 
                                 subsample=1.0, 
                                 criterion='friedman_mse', 
                                 min_samples_split=2, 
                                 min_samples_leaf=1, 
                                 min_weight_fraction_leaf=0.0, 
                                 max_depth=7, 
                                 min_impurity_decrease=0.0, 
                                 min_impurity_split=None, 
                                 init=None, random_state=None, 
                                 max_features=None, 
                                 verbose=0, 
                                 max_leaf_nodes=None, 
                                 warm_start=False, 
                                 presort='auto')




fit = rgf.fit(train.data, train.target)

'''rgf_scores = cross_val_score(rgf,
                             train.data,
                             train.target,
                             cv=StratifiedKFold(n_folds))
'''
#rgf_score = sum(rgf_scores)/n_folds
#print('RGF Classfier score: {0:.5f}'.format(rgf_score))
print('done')



'''
b = fit.predict(train.data)
print(sum(b))


print([b[i] for i in range(len(b))])

rgf_scores = cross_val_score(rgf,
                             train.data,
                             train.target,
                             cv=StratifiedKFold(3))
print(rgf_scores)
print('done')



from sklearn import cross_validation,metrics
from sklearn import svm

train_x,test_x,train_y,test_y = cross_validation.train_test_split(train.data,train.target,test_size=0.2,random_state=27)#把训练集按0.2的比例划分为训练集和验证集
#start svm
rgf.fit(train_x,train_y)
predict_prob_y = rgf.predict_proba(test_x)#基于SVM对验证集做出预测，prodict_prob_y 为预测的概率
#end svm ,start metrics 
test_auc = metrics.roc_auc_score(test_y,predict_prob_y[:,1])#验证集上的auc值
print (test_auc)
'''
```
